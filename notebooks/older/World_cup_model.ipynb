## The models

### Teams

```python
class Team:
    __slots__ = ['name', 'atts', 'defs', '_hash']
    
    def __init__(self, name: str, atts: float = 0, defs: float = 0):
        self.name = name
        self.atts = atts
        self.defs = defs
        self._hash = hash((self.name, self.atts, self.defs))
    
    def __hash__(self):
        return self._hash
        
    def __str__(self):
        return f"{self.name} - atts: {self.atts}, defs: {self.defs}"
    
    def __repr__(self):
        return f"Team('{self.name}', {self.atts}, {self.defs})"
    
    def __eq__(self, other):
        if isinstance(other, self.__class__):
            return hash(self) == hash(other)
        else:
            return False

if TEST:
    nor = Team(name='Norway', atts=.2, defs=.1)
    bra = Team('Brazil', atts=.8, defs=.5)
    den = Team('Denmark', atts=.4, defs=.15)
    aus = Team('Australia', atts=.6, defs=.3)
    print(nor)
```

### Matches

```python
stats = collections.namedtuple('stats', 'team goals_scored goals_admitted points')

class Match:
    def __init__(self, home: Team, away: Team, can_draw: bool = True, intercept:float = 0, home_advantage:float = 0):
        assert isinstance(home, Team), f"home team has to be a Team instance, instead {repr(home)}"
        assert isinstance(away, Team), f"away team has to be a Team instance, instead {repr(away)}"
        self.home = home
        self.away = away
        self.intercept = intercept
        self.home_advantage = home_advantage
        self.can_draw = can_draw
        self.home_goals = None
        self.away_goals = None
        self.to_overtime = False
        
        self.home_theta = np.exp(self.intercept + self.home.atts + self.away.defs + self.home_advantage )
        self.away_theta = np.exp(self.intercept + self.away.atts + self.home.defs)
        
        self.play()
    
    def __repr__(self):
        return f"Match({repr(self.home)}, {repr(self.away)}, {self.can_draw}, {self.intercept}, {self.home_advantage})"
    
    def play(self):
        # Ordinary time
        home_goals = np.random.poisson(self.home_theta)
        away_goals = np.random.poisson(self.away_theta)

        if (home_goals == away_goals) and (not self.can_draw): # Assumes golden goal
            self.to_overtime = True
            first_home_goal = np.random.exponential(1/self.home_theta)
            first_away_goal = np.random.exponential(1/self.away_theta)
            if first_home_goal < first_away_goal:
                home_goals += 1
            else:
                away_goals += 1
        self.home_goals = home_goals
        self.away_goals = away_goals
        if home_goals > away_goals:
            self.winner, self.loser = self.home, self.away
        elif away_goals > home_goals:
            self.winner, self.loser = self.away, self.home
        else:
            self.winner = None
        
    @property
    def stats(self):
        try:
            winner = self.winner.name
        except AttributeError:
            winner = 'draw'
        return {self.home.name: self.home_goals, self.away.name: self.away_goals, 'stats': {'winner': winner, 'can_draw': self.can_draw, 'to_overtime': self.to_overtime}}
    
    @staticmethod
    def _score_points(own_goals, other_goals):
        if own_goals > other_goals:
            return 3
        elif own_goals < other_goals:
            return 0
        return 1
    
    @property
    def teams(self):
        return [self.home, self.away]
    
    @property
    def home_stats(self):
        return stats(repr(self.home), self.home_goals, self.away_goals, self._score_points(self.home_goals, self.away_goals))
    
    @property
    def away_stats(self):
        return stats(repr(self.away), self.away_goals, self.home_goals, self._score_points(self.away_goals, self.home_goals))

if TEST:
    m = Match(home=nor, away=bra, intercept=.2)
    print(m.stats)
    print(m.away_stats)
```

### Groups

Read advancement rules here if in doubt: https://www.uefa.com/MultimediaFiles/Download/Regulations/uefaorg/Regulations/01/87/54/21/1875421_DOWNLOAD.pdf

```python
class Group:
    def __init__(self, teams: List[Team], intercept: float = 0, home_advantage: float = 0):
        self.teams = teams
        self.intercept = intercept
        self.home_advantage = home_advantage
        self.games = self.gen_fixture()
        self.ranking = self.play()
        
    
    def gen_fixture(self):
        team_list = []
        for team1 in self.teams:
            for team2 in self.teams:
                if (team2 != team1) and ((team2, team1) not in team_list):
                    team_list.append((team1, team2))
        return [Match(team1, team2, can_draw=True, intercept=self.intercept, home_advantage=self.home_advantage) for (team1, team2) in team_list]
    
    def play(self):
        results = []
        for g in self.games:
            results.append(g.home_stats)
            results.append(g.away_stats)
        self.results = pd.DataFrame(results).assign(goals_diff = lambda df: df.goals_scored - df.goals_admitted)
        
        
        return (self.results
                .groupby('team')
                .sum()
                .sort_values(by=['points', 'goals_scored', 'goals_diff'], ascending=[False, False, False])
                [['points', 'goals_scored', 'goals_diff', 'goals_admitted']]
                .reset_index()
               )
    
    @property
    def winner(self):
        return eval(self.ranking.loc[0, 'team'])
   
    @property
    def second(self):
        return eval(self.ranking.loc[1, 'team'])
        
    
if TEST:        
    g = Group([nor, bra, den, aus])
    print(g.games)
    print(g.ranking)
    print(f'The group winner is {g.winner}')
```

### Tournament

```python
class Playoffs:
    """
    The World Cup 2018 playoff structure. To be used both in WC simulations, and when the WC is underway.
    The team dict has to have keys "A1" etc for every group (A, ..., H) and 1 and 2. Values are Team instances.
    """
    def __init__(self, team_dict, intercept: float = 0, home_advantage: float = 0):
        necessary_keys = [''.join([el[0], el[1]]) for el in itertools.product(list('ABCDEFGH'), list('12'))]
        for key in necessary_keys:
            assert isinstance(team_dict[key], Team)
        self.team_dict = team_dict
        self.intercept = intercept
        self.home_advantage = home_advantage
        
        self.play()
        
    def play(self):
        playoffs = {}
        # 1/8 finals
        playoffs[49] = Match(self.team_dict['D2'], self.team_dict['C1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[50] = Match(self.team_dict['B2'], self.team_dict['A1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[51] = Match(self.team_dict['A2'], self.team_dict['B1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[52] = Match(self.team_dict['C2'], self.team_dict['D1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[53] = Match(self.team_dict['F2'], self.team_dict['C1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[54] = Match(self.team_dict['H2'], self.team_dict['G1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[55] = Match(self.team_dict['E2'], self.team_dict['F1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[56] = Match(self.team_dict['G2'], self.team_dict['H1'], intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        
        # 1/4 finals
        playoffs[57] = Match(playoffs[50].winner, playoffs[49].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[58] = Match(playoffs[54].winner, playoffs[53].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[59] = Match(playoffs[56].winner, playoffs[55].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[60] = Match(playoffs[52].winner, playoffs[51].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        
        # Semi finals
        playoffs[61] = Match(playoffs[58].winner, playoffs[57].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        playoffs[62] = Match(playoffs[60].winner, playoffs[59].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        
        # Bronze final
        playoffs[63] = Match(playoffs[62].loser, playoffs[61].loser, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        
        # Final
        playoffs[64] = Match(playoffs[62].winner, playoffs[61].winner, intercept=self.intercept, home_advantage=self.home_advantage, can_draw=False)
        
        self.playoffs = playoffs
        
        self.winner = playoffs[64].winner.name

    def get_placement(self, team):
        if team == self.playoffs[64].winner:
            return 1
        elif team == self.playoffs[64].loser:
            return 2
        elif team == self.playoffs[63].winner:
            return 3
        elif team == self.playoffs[63].loser:
            return 4
        elif team in list(itertools.chain.from_iterable([self.playoffs[x].teams for x in [57, 58, 59, 60]])):
            return 8
        elif team in list(itertools.chain.from_iterable([self.playoffs[x].teams for x in [49, 50, 51, 52, 53, 54, 55, 56]])):
            return 16
        else:
            return None


        
```

```python
class WorldCup:
    """
    Simulates the world cup. Requires a team_list, which is a list of 32 Team instances. 
    Every nation that is in the World Cup has to be represented here.
    """
    def __init__(self, team_list: List[Team], intercept: float = 0, home_advantage: float = 0):
        self.grouping = json.loads(Path('../data/groups.json').read_text())
        self.control_entries(team_list)
        self.team_list = team_list
        self.intercept = intercept
        self.home_advantage = home_advantage
        
        self.play()
        
    
    def control_entries(self, team_list):
        assert len(team_list) == 32, "Must enter 32 teams"
        for team in team_list:
            assert isinstance(team, Team), "All teams must be Team instances"
        # Teams      
        required_teams = sorted([team for lst in self.grouping.values() for team in lst])
        entered_teams = [team.name for team in team_list]
        for team in required_teams:
            assert team in entered_teams, f"team_list lacks team {team}"
        return True
    
    def group_play(self):
        groups = {}
        for group, teams in self.grouping.items():
            group_teams = []
            for team in teams:
                for entry in self.team_list:
                    if entry.name == team:
                        group_teams.append(entry)
            groups[group] = Group(group_teams, intercept=self.intercept, home_advantage=self.home_advantage)
        self.groups = groups

    def play(self):
        self.group_play()

        self.playoffs = Playoffs(self.advancement, intercept=self.intercept, home_advantage=self.home_advantage)
               
    @property
    def advancement(self):
        adv = {}
        for group in self.groups.keys():
            adv[group+'1'] = self.groups[group].winner
            adv[group+'2'] = self.groups[group].second
        return adv
    
    def get_placement(self, team):
        place = self.playoffs.get_placement(team)
        if place is not None:
            return place
        return 24
        
    @property
    def stats(self):
        outcome = [(self.get_placement(team), team.name) for team in self.team_list]
        outcome = pd.DataFrame(outcome, columns=['rank', 'team']).sort_values(by='rank').set_index('team', drop=True)
        return outcome

                
        
if TEST:            
    entry_teams = [Team(team_name, np.random.random(), np.random.random()) for team_name in entering_teams]
    wc = WorldCup(entry_teams)
    print(wc.stats)
```

```python
def get_team_idx(team_name):
    return teams.loc[teams['team'] == team_name, 'team_idx'].values[0]
```

```python
def make_draw(seed=None, russian_home_advantage=True):
    np.random.seed(seed)
    idx = np.random.choice(trace['intercept'].shape[0])
    intercept = trace['intercept'][idx]
    home_advantage = trace['home'][idx]
    team_list = []
    for team in entering_teams:
        atts = trace['atts'][idx, get_team_idx(team)]
        defs = trace['defs'][idx, get_team_idx(team)]
        if (team == 'Russia') and russian_home_advantage:
            atts += home_advantage
        team_list.append(Team(name=team, atts=atts, defs=defs))
    return dict(team_list=team_list, intercept=intercept, home_advantage=0)

def median_params(russian_home_advantage=True):
    intercept = pm.stats.quantiles(trace['intercept'])[50]
    home_advantage = pm.stats.quantiles(trace['home'])[50]
    atts = pm.stats.quantiles(trace['atts'])[50]
    defs = pm.stats.quantiles(trace['defs'])[50]
    team_list = []
    for team in entering_teams:
        atts_team = atts[get_team_idx(team)]
        defs_team = defs[get_team_idx(team)]
        if (team == 'Russia') and russian_home_advantage:
            atts_team += home_advantage
        team_list.append(Team(name=team, atts=atts_team, defs=defs_team))
    return dict(team_list=team_list, intercept=intercept, home_advantage=0)

if TEST:
    print(WorldCup(**median_params()).stats.T)
```

```python
class Simulation:
    """
    Class for simulating the World Cup 2018 with different parameters
    """
    def __init__(self, pars: list, n: int = 1):
        """
        Runs n simulations for each set of parameters in pars, and returns the aggregate statistics
        """
        self.pars = pars
        self.n = n
        self.sim = self.simulate()
        
    def simulate(self):
        out = pd.concat([WorldCup(**par).stats.T for par in self.pars for _ in range(self.n)], axis=0).reset_index(drop=True)
        return (pd
                 .concat([
                     out.describe().T.sort_values(by='50%'),
                     out.agg(lambda col: np.mean(col==1)).rename('share wins'),
                     out.agg(lambda col: np.mean(col<=4)).rename('share top 4'),
                     out.agg(lambda col: np.mean(col<24)).rename('share playoff')
                         ], axis=1)
                 .drop('count', axis='columns')
                 .merge(right=pd.DataFrame([(group, country) for group in groups.keys() for country in groups[group]], columns = ['group', 'country']),
                        left_index=True,
                        right_on='country'
                       )
                 .sort_values(by='share wins', ascending=False)
                 .reset_index(drop=True)
                 [['country', 'group', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'share wins', 'share top 4', 'share playoff']]
                )

    @property
    def group(self):
        colors = []
        for _, color in menon_styles.menon_farger.items():
            colors.append(tuple([int(255*el) for el in color]))

        return (self.sim
                .sort_values(by=['group', 'share playoff'], ascending=[True, False])
                .reset_index(drop=True)
                .style
                 .apply(lambda row: [f"background-color: {'#%02x%02x%02x'%colors[(ord(row.group)-1)%8]}"]*len(row), axis=1)
                 #.bar(subset=['share wins', 'share top 4', 'share playoff'], align='left', color=['#5fba7d'])
                 .format({'share wins': "{:.0%}", 'share top 4': "{:.0%}", 'share playoff': "{:.0%}"})
               )
    
    @property
    def style(self):
        return (self.sim
                .style
                .bar(subset=['share wins', 'share top 4', 'share playoff'], align='left', color=['#5fba7d'])
                .format({'share wins': "{:.0%}", 'share top 4': "{:.0%}", 'share playoff': "{:.0%}"})
                .set_properties(**{'border-color': 'black'})
               )
```
